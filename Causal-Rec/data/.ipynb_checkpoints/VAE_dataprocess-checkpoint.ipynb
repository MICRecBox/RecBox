```python
import  pandas as pd
import numpy as np
import os
from tqdm import  tqdm
from sklearn.preprocessing import  LabelEncoder
import json
from joblib import  Parallel,delayed
from functools import  partial
from matplotlib import  pyplot as plt
import random

path = 'E:\data'
savePath = os.getcwd()
print(f'path is {path} \n savePath is {savePath}')
```

```python
#REASONER
data = pd.read_csv(path + '\REASONER-Dataset\interaction.csv',sep='\t',header = 0)[['user_id','video_id','like','rating']]
print(data.shape)
data.head()
```

```python
a = data.groupby('rating').like.count().tolist()
b = data.groupby('rating').like.sum().tolist()
c = np.array(b) / np.array(a)
print(a,b,c)
labels = [1,2,3,4,5]
x = np.arange(len(labels))
width = 0.35
plt.figure(figsize=(6,4),dpi = 300)
plt.bar(x - width / 2,a,width,label = 'rating')
plt.bar(x + width / 2,b,width, label = 'like')
plt.show()
```
